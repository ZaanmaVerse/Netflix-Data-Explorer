# ðŸ“Š Netflix Data Explorer
Eksplorasi data Netflix menggunakan Python dan Pandas.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Pengaturan tampilan visual
sns.set(style='darkgrid')
```

## 1. Load Dataset

```python
df = pd.read_csv("../data/netflix_titles.csv")
df.head()
```

## 2. Informasi Umum Dataset

```python
df.info()
```

## 3. Jumlah Film vs TV Shows

```python
sns.countplot(data=df, x='type')
plt.title('Jumlah Konten berdasarkan Tipe')
plt.xlabel('Type')
plt.ylabel('Jumlah')
plt.show()
```

## 4. Distribusi Tahun Rilis

```python
df['release_year'].value_counts().sort_index(ascending=False).head(10).plot(kind='bar', figsize=(10,5), color='skyblue')
plt.title('Jumlah Konten Berdasarkan Tahun Rilis (Top 10)')
plt.xlabel('Tahun')
plt.ylabel('Jumlah')
plt.show()
```

## 5. Negara Produksi Teratas

```python
df['country'].value_counts().head(10).plot(kind='barh', figsize=(8,5), color='coral')
plt.title('Top 10 Negara Produksi')
plt.xlabel('Jumlah')
plt.ylabel('Negara')
plt.show()
```

## 6. Genre / Kategori Populer

```python
from collections import Counter
from itertools import chain

genres = df['listed_in'].dropna().apply(lambda x: [g.strip() for g in x.split(',')])
genre_counts = Counter(chain.from_iterable(genres))
pd.Series(genre_counts).sort_values(ascending=False).head(10).plot(kind='bar', color='green')
plt.title('Top 10 Genre Konten Netflix')
plt.ylabel('Jumlah')
plt.xticks(rotation=45)
plt.show()
```

## 7. Sutradara Terpopuler

```python
df['director'].value_counts().head(10).plot(kind='barh', color='purple')
plt.title('Top 10 Sutradara Terpopuler')
plt.xlabel('Jumlah')
plt.ylabel('Sutradara')
plt.show()
```

## 8. Pemeran Populer

```python
actors = df['cast'].dropna().apply(lambda x: [a.strip() for a in x.split(',')])
actor_counts = Counter(chain.from_iterable(actors))
pd.Series(actor_counts).sort_values(ascending=False).head(10).plot(kind='bar', color='orange')
plt.title('Top 10 Pemeran Populer di Netflix')
plt.xticks(rotation=45)
plt.ylabel('Jumlah')
plt.show()
```
